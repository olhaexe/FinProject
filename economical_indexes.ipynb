Dow Jones
Промышленный индекс Доу-Джонса (DJIA) показывает изменения котировок акций 30 крупнейших корпораций США. Каждая компания в составе Индекса оказывает на индекс одинаковое влияние независимо от своей капитализации. Компоненты индекса периодически пересматриваются редакцией Wall Street Journal. Индекс рассчитывается с 1884 года, и впервые был опубликован 26 мая 1896 года. Данный индекс прекрасно подходит для измерения динамики котировок крупнейших по капитализации компаний рынка, аналогично европейским индексам CAC-40 или DAX, однако для измерения динамики широкого рынка США лучше подходит S&P 500.


S&P 500
S&P-500 - взвешенный по рыночной капитализации (free - float) индекс акций 500 корпораций, которые представлены в нем в следующей пропорции: 400 промышленных корпораций, 20 транспортных, 40 финансовых и 40 коммунальных компаний. Данный индекс часто называют «барометром американской экономики». Акции из списка S&P-500 котируются на крупных американских фондовых биржах. В качестве базового периода для расчёта взяты 1941—1943 года, базовое значение — 10.

NASDAQ Composite
NASDAQ Composite Index - сводный взвешенный по капитализации индекс, который включает в себя более 3000 ценных бумаг высокотехнологичных и быстро растущих компаний, котирующихся на фондовой бирже Nasdaq. В отличие от других рыночных индексов, Nasdaq Composite не ограничивается компаниями, базирующимися в США. Индекс рассчитывается на основе средневзвешенных цен непрерывно в течение всего торгового дня.

Nikkei 225
Nikkei 225 (Nikkei Stock Average) — ценовой взвешенный по рыночной капитализации индекс, включающий акции 225 наиболее активно торгуемых компаний первой секции Токийской фондовой биржи. Nikkei полностью отображает ситуацию на фондовом рынке Японии и в экономике страны в целом. Впервые Индекс был опубликован 7 сентября 1950 года Токийской фондовой биржей под названием «TSE Adjusted Stock Price Average».

KOSPI
KOSPI - основной фондовый индекс Южной Кореи, рассчитанный на основе взвешенных по рыночной капитализации цен всех торгующихся на Корейской фондовой бирже обыкновенных акций (привилегированные акции исключены из расчета с 14 июня 2002 года). Индекс начал рассчитываться в 1983 году, заменив собой индекс KCSPI (Korea Composite Stock Price Index - Корейский индекс цен на сырьевые товары). База расчета Индекса составляет 100 пунктов на 4 января 1980 года.
VIX Index
The CBOE Volatility Index® (VIX®) is a key measure of market expectations of near-term volatility conveyed by S&P 500 stock index option prices. Since its introduction in 1993, VIX has been considered by many to be the world’s premier barometer of investor sentiment and market volatility.

MSCI World Index
The MSCI World Index is a broad global equity index that represents large and mid-cap equity performance across all 23 developed markets countries. It covers approximately 85% of the free float-adjusted market capitalization in each country.

S&P/TSX Composite
The S&P/TSX Composite Index has provided investors with a premier indicator of market activity for Canadian equity markets since its launch in 1977. With approximately 95% coverage of the Canadian equities market, it is the primary gauge for Canadian-based, Toronto Stock Exchange listed companies. The S&P/TSX Composite Index serves the dual purpose of a benchmark and an investable index. The index is designed to offer the representation of a broad benchmark index while maintaining the liquidity characteristics of narrower indices.

Shiller PE Ratio (CAPE ratio)
Коэффициент Shiller P/E был предложен и разработан профессором Йельского университета Робертом Шиллером, автором книги Irrational Exuberance (“Иррациональный оптимизм”). Методика его расчета предполагает корректировку значений текущих цен акций (Рrice, P) или индекса (например, S&P 500) и их прибылей (Earning, E) с поправкой на инфляцию за последние 10 лет (CPI). Предполагается, что такой подход сглаживает цикличность, присутствующую в бизнесе компаний, и более точно отслеживает “реальные” (с поправкой на инфляцию) значения индекса широкого рынка S&P 500.

ASX 200
S&P/ASX 200 является институциональным инвестиционным ориентиром в Австралии. S&P/ASX 200 предназначен для расчета производительности 200 крупнейших акций, котирующихся на ASX, основываясь на рыночной капитализации с поправкой на свободное обращение.

## Выбросы

```python
boxplot_list = []

for col in df2.columns[1:]:
    q1n = df[col].quantile(0.25)
    q3n = df[col].quantile(0.75)
    iqrn = q3n - q1n
    boxplot_range_n = (q1n - 1.5 * iqrn, q3n + 1.5 * iqrn)
    df_boxplot = df2[(df2[col] < boxplot_range_n[0]) | (df2[col] > boxplot_range_n[1])]
    if df_boxplot.shape[0] != 0:
        print(f'{col}: {df_boxplot.shape[0]} elements')
        boxplot_list.append(col)
        #print(df_boxplot)
```

## Отбор и построение новых признаков

```python
# Предполагаю, что в сложные периоды высокая волатильность и будет большая разница между минимальным и максимальным значением,
# а также между открытием и закрытием. Построим признаки там, где есть эта информация.
df2['delta_btc_mm'] = df2['Btc_max'] - df2['Btc_min']
df2['delta_btc_oc'] = df2['Btc_open'] - df2['Btc_close']
df2['delta_moex_mm'] = df2['MOEX_max'] - df2['MOEX_min']
df2['delta_moex_oc'] = df2['MOEX_open'] - df2['MOEX_close']
df2['delta_rts_mm'] = df2['RTS_max'] - df2['RTS_min']
df2['delta_rts_oc'] = df2['RTS_open'] - df2['RTS_close']

df2.head(2)
```
